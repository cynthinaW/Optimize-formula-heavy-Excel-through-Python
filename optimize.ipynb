```python
import pandas as pd
import datetime
```

```python
dat1 = pd.read_excel('test.xlsx',
                     sheet_name = 'Inventory')
dat2 = pd.read_excel('test.xlsx',
                     sheet_name = 'All Shipments')
dat3 = pd.read_excel('test.xlsx',
                     sheet_name = 'Exclusions_shipment')
dat4 = pd.read_excel('test.xlsx',
                     sheet_name = 'Exclusions_SKU')
```

```python
dat2
```

```python
dat3.rename(columns={'SHIPMENT':'shipment_id', 'REASON':'reason_shipment'}, inplace = True)
dat4.rename(columns={'SKU':'sku', 'REASON':'reason_sku'}, inplace = True)
dat3
```

```python
dat1['Docks'] = dat1.apply(lambda x: x['Dock Transit To Reno'] + x['Reno_Staging'] + x['Reno'] + x['Indy_Stage'] + 
                          x['Indy_Transit'] + x['Indy'] + x['Indy_Putaway'] + x['BK_Docks'] + x['Brooklyn'], axis = 1)
```

```python
shipment_id = [int(i) for i in list(dat2['shipment_id'])]
sku = [int(i) for i in list(dat2['sku'])]
quantity = [int(i) for i in list(dat2['quantity'])]
#important
datt = pd.DataFrame({'shipment_id':shipment_id, 'sku':sku, 'quantity':quantity})

datt['code'] = datt.apply(lambda x: str(x['shipment_id'])+'-'+str(x['sku']), axis = 1)
```

```python
datt
```

```python
dat3
```

```python
dat3.loc[:,['shipment_id', 'reason_shipment']]
```

```python
datt1 = pd.merge(datt, dat3.loc[:,['shipment_id', 'reason_shipment']], how = 'left', on = 'shipment_id')
datt2 = pd.merge(datt1, dat4.loc[:,['sku', 'reason_sku']], how = 'left', on = 'sku')
datt = datt2
datt
```

```python
dat1
```

```python
datt1 = pd.merge(dat2.groupby(['sku']).sum().reset_index().loc[:,['sku', 'quantity']], 
                 dat1.groupby(['sku']).sum().reset_index().loc[:,['sku', 'Docks']], how = 'left', on = 'sku')

datt1 = datt1[datt1['Docks'].notna()]
datt1['minus'] = datt1.apply(lambda x:x['Docks'] - x['quantity'], axis = 1)

l = [] 
for i in list(datt1['minus']):
    if i == 0:
        l.append('Just Enough')
    elif i < 0:
        l.append('No')
    else:
        l.append('Yes')
datt1['status'] = l
datt1
del datt1['quantity']
del datt1['Docks']
datt1 = datt1[datt1['status'] != 'Yes']
datt1
```

```python
datt
```

```python
datt_tmp = pd.merge(datt, datt1, how = 'left', on = 'sku')
datt = datt_tmp
```
